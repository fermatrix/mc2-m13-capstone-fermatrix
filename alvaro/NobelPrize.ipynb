```python
new_types={
        "id":"int",
        "fullName":"string",
        "fileName":"string",
        "gender":"string",
        "birth":"string"
    }


df_laureates=pd.read_parquet("laureates-1901-2019.parquet",  use_pandas_metadata=True)
df_laureates=df_laureates.astype(new_types)
df_laureates.dropna(subset=['fullName'],inplace=True)
df_laureates=df_laureates.drop_duplicates(subset=['id','fullName','fileName'])
df_laureates=df_laureates.set_index('id')
df_laureates.sort_index(inplace=True)

```

```python
df_nobel_prizes = pd.read_parquet('nobelPrizes-1901-2019.parquet')

df_nobel_prizes.dropna(subset = ['dateAwarded'], inplace = True)
df_nobel_prizes.head(60)


new_types = {
    'awardYear':'string',
    'category':'string',
    'categoryFullName':'string',
    'dateAwarded':'string',
    'prizeAmount':'int',
    'prizeAmountAdjusted':'int',
    'laureates_id':'int',
}



df_laureates.dropna(subset = ['fullName'], inplace = True)
df_laureates.head(60)
df_nobel_prizes=df_nobel_prizes.explode('laureates_id')

df_nobel_prizes=df_nobel_prizes.astype(new_types)

df_nobel_prizes=df_nobel_prizes.set_index('laureates_id')

mix_df = df_nobel_prizes.join(df_laureates, on='laureates_id', how='inner', lsuffix='_df_laureate', rsuffix='_df_nobel')
df_nobel_prizes.head(60)
# mix_df.head(60)
```
